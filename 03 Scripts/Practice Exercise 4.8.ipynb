Seems like nothing has happened, but it has worked. You just can't see it. Groupby function by itself is useless (you need to do something else to it) Needs to be used as a series of steps:

1. Split data into groups based on some criteria
2. Apply function to each group separately
3. Combine results into dataframe or create new column in current dataframe

# Aggregating Data with agg()

These functions create summaries of selected columns and store the summary values in new columns

```python
df.groupby('department_id').agg({'order_number': ['mean']})
```

```python
# Some aggregations can be conducted without use of agg() function
df.groupby('department_id')['order_number'].mean()
```

Key difference in syntax between the 2 methods is when using agg() function, put column you want to aggregate inside (). When using mean() function, index the column with [] instead and follow it with function you want to use after "."

** Good idea to use square brackets bc they have no other role beyond indexing, "." have a role in Python.

```python
# Performing multiple aggregations
df.groupby('department_id').agg({'order_number': ['mean', 'min', 'max']})
```

# Aggregating data with transform()

```python
df_ords_prods_merge['max_order'] = df_ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

# Deriving columns with loc()

```python
# Create flag that assigns "loyalty" label to user ID based on corresponding max order value
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['max_order'] <= 40) & (df_ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
df_ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```
